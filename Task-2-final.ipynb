# Use Clustering Techniques for the any customer dataset using machine learning

```python
# Load the Data:
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
# Check for Missing Values:
```

```python
missing_values = df.isnull().sum()
print("Missing Values:")
print(missing_values)
```

```python
# Encode Categorical Variables:
```

```python
df_encoded = pd.get_dummies(df, columns=['Genre'], drop_first=True)
```

```python
# Feature Scaling:
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
features = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
df_encoded[features] = scaler.fit_transform(df_encoded[features])
```

```python
#Visualization:
```

```python
# Data Exploration:
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.pairplot(df_encoded, hue='Genre_Male')
plt.show()
```

```python
# Clustering with K-Means:
```

```python
# Train the K-Means Model:
```

```python
from sklearn.cluster import KMeans

k = 3  # Adjust the number of clusters (K)
kmeans_model = KMeans(n_clusters=k, random_state=42)
kmeans_model.fit(df_encoded)
```

```python
# Assign Clusters to Data Points:
```

```python
df_encoded['Cluster'] = kmeans_model.labels_
```

```python
# Visualization of Clusters:
```

```python
plt.scatter(df_encoded['Annual Income (k$)'], df_encoded['Spending Score (1-100)'], c=df_encoded['Cluster'], cmap='rainbow')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('Cluster Visualization')
plt.show()
```
